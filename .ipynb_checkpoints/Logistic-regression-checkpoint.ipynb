### Step 1 : 数学的な基礎

```python
# ロジスティック関数
def logistic(t):
    return 1.0/(1+math.exp(-1.0*t))
```

```python
# tを−６から6まで500点作成
t = np.linspace(-6,6,500)
```

```python
y = np.array([logistic(ele) for ele in t])
```

```python
plt.plot(t, y)
plt.title('Logistic Function')
```

### Step 2 : データの準備/可視化

```python
df = sm.datasets.fair.load_pandas().data
```

```python
df.head()
```

```python
# 不倫の有無を判定する関数を定義
def affair_check(x):
    if x != 0:
        return 1
    else:
        return 0
```

```python
df['Had_Affair'] = df['affairs'].apply(affair_check)
```

```python
df
```

```python
# 不倫の有無でグループ化して平均値を算出
df.groupby('Had_Affair').mean()
```

不倫したことがある女性の傾向？
・年齢が高い
・結構期間が長い
・宗教観があまりない
・学歴が低い

```python
# 年齢でプロット
sns.countplot('age',data=df.sort_values('age'),hue='Had_Affair', palette='coolwarm')
```

```python
# 結婚期間でプロット
sns.countplot('yrs_married',data=df.sort_values('yrs_married'),hue='Had_Affair', palette='coolwarm')
```

```python
# 子供の数でプロット
sns.countplot('children',data=df.sort_values('children'),hue='Had_Affair', palette='coolwarm')
```

```python
# 学歴でプロット
sns.countplot('educ',data=df.sort_values('educ'),hue='Had_Affair', palette='coolwarm')
```

### Step 3 : データの前処理

```python
# カテゴリーを表現する変数をダミー変数に展開
occ_dummies = pd.get_dummies(df['occupation'])
hus_occ_dummies = pd.get_dummies(df['occupation_husb'])
```

```python
occ_dummies.head()
```

```python
# 列名を付ける
occ_dummies.columns = ['occ1','occ2','occ3','occ4','occ5','occ6']
hus_occ_dummies.columns = ['hocc1','hocc2','hocc3','hocc4','hocc5','hocc6']
```

```python
# occupationは不要
# Had_Affairも目的変数のため不要
X = df.drop(['occupation','occupation_husb','Had_Affair'],axis=1)
```

```python
# ダミー変数のDataFrameを繋げる
dummies = pd.concat([occ_dummies, hus_occ_dummies], axis=1)
```

```python
# すべてを連結して説明変数のDataFrameを作成
X = pd.concat([X,dummies], axis=1)
```

```python
X.head()
```

```python
# Yに目的変数を格納
Y = df.Had_Affair
```

```python
Y.head()
```

```python
Y.tail()
```

### Step 4 : 多重共線性

```python
# 多重共線性の問題を回避
X = X.drop('occ1', axis=1)
X = X.drop('hocc1', axis=1)

# 目的変数の元になっている変数なので削除
X = X.drop('affairs', axis=1)
```

```python
X.head()
```

```python
Y = Y.values
# Y = np.ravel(Y)でも可
```

### Step 5 : ScikitLearnを使ったロジスティック回帰

```python
log_model = LogisticRegression()
```

```python
# モデルの作成
log_model.fit(X, Y)
```

```python
# モデルの精度を確認
log_model.score(X,Y)
```

```python
Y.mean()
```

```python
# 変数名とその係数を格納するDataFrameを作成
coeff_df = DataFrame([X.columns, log_model.coef_[0]]).T
```
